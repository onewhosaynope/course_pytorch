**TASK: Using pandas .apply() method, create a new column called "marital code". This column will only contained a shortened code of the possible marital status first letter. (For example "m" for "married" , "s" for "single" etc... See if you can do this with a lambda expression. Lots of ways to do this one!**

[Hint Link](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html)


**ЗАДАЧА: Используя метод pandas .apply (), создайте новый столбец под названием «marital code». Этот столбец будет содержать только сокращенный код первой буквы возможного семейного положения. (Например, "m" для "married", "s" для "single" и т. Д. Посмотрим, сможете ли вы сделать это с помощью лямбда-выражения. Есть много способов сделать это!**

[Ссылка-подсказка](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html)

```python
def marital_status(status):
    if status == 'married':
        return 'm'
    elif status == 'single':
        return 's'
    else:
        return 't'

data_set['marital code'] = data_set['marital'].apply(marital_status) # or .apply(lambda status: status[0])
data_set.head()
```
